```python
import os
from dotenv import load_dotenv
from langchain.chat_models import ChatOpenAI
from langchain import PromptTemplate
from langchain.chains import LLMChain
from langchain.schema import (
    AIMessage,
    HumanMessage,
    SystemMessage
)

load_dotenv()

# api_key = os.getenv("OPENAI_API_KEY")
api_key = "REDACTED_OPENAI_API_KEY"
api_key = "REDACTED_OPENAI_API_KEY"
```

```python
def read_trans_code(transfolder, codefolder, name, extension):
    transfile = transfolder + name + "_transcript.json"
    codefile = codefolder + name + "_code." + extension
    with open(transfile) as f:
        transcript =  f.read()
    with open(codefile) as f:
        code = f.read()
    return transcript, code
    
```

```python
transfolder = "./neetcode/"
codefolder = "./codes/neetcode/"
transcript, code = read_trans_code(transfolder, codefolder, "1pkOgXD63yU", "py")
```

```python
template = """
I have a video transcript where an instructor is teaching the solution of a leetcode problem, I also have their code solutions.
### Transcript
{transcript}
### Code
{code}
### Instruction: please align the text with the code that they are referring to, please output in a json format with three fields: `timestamp`, `transcript`, and `code`.
"""
```

```python
from langchain_core.pydantic_v1 import BaseModel
class Alignment(BaseModel):
    timestamp: str
    transcript: str
    code: str
```

```python
chat_model = ChatOpenAI(openai_api_key=api_key, model_name="gpt-4o",temperature=0)
```

```python
from langchain_core.output_parsers import JsonOutputParser

# Set up a parser + inject instructions into the prompt template.
parser = JsonOutputParser(pydantic_object=Alignment)

prompt = PromptTemplate(
    template=template,
    input_variables=["transcript", "code"],
    partial_variables={"format_instructions": parser.get_format_instructions()},
)

chain = prompt | chat_model | parser

chain.invoke({"transcript": transcript, "code": code})
```
